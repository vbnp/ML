```python
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[1,2],[2,2],[2,3],[8,7],[8,8],[25,80]])
```

```python
x
```

```python
x[:,0]
```

```python
x[:,1]
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
db = DBSCAN(eps =3,min_samples = 2)
db.fit(x)
db.labels_
```

```python
db = DBSCAN(eps =3,min_samples = 2)
clusters =db.fit_predict(x)
plt.scatter(x[:,0],x[:,1],c=clusters,cmap = 'viridis',marker = 'o')
```

```python
from sklearn.datasets import make_circles
```

```python
x1,_ = make_circles(n_samples=500,factor=.5,noise = .03,random_state=4)
```

```python
x1[:,0]
```

```python
db1=DBSCAN(eps=0.1,min_samples=5)
clusters = db1.fit_predict(x1)
plt.scatter(x1[:,0],x1[:,1], c=clusters,cmap='viridis',marker = 'o')
plt.title("DBSCAN clustering of Concentric Circle")
plt.xlabel("Feature 0")
plt.ylabel('Feature 1')
plt.show()
```
